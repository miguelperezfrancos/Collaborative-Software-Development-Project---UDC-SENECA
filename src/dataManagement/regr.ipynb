```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

## Provide our model the data

```python
df = pd.read_csv('/Users/rafa/Documents/Uni/Cuatri3/software_engineering/sample_data/housing.csv')
cols = df.columns
cols
```

### Set target variable (y) and feature variable (x)

```python
x = df[['median_income']]
y = df['median_house_value']

print(f'max median value: {y.max()}')
```

### Create the model and fit it with our variables

```python
# we create the model
model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(X=x)
```

### Measure the error of our model

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

# Print the error measures
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'R-squared (R²): {r2}')
```

### Get the regression line definition

```python
slope = model.coef_[0]
intercept = model.intercept_

print(f'Regression line: y = {slope:.2f} * x + {intercept:.2f}')

regression_line = lambda x: slope*x + intercept
```

### Plot the output

```python
# Plot the result of the linear regression

plt.scatter(x, y, color='blue', label='Data points', s=1)
plt.plot(x, y_pred, color='red', label='Regression line')
plt.plot(x, regression_line(x), color='green', label='line definition')
plt.xlabel('Median income')
plt.ylabel('median house value')
plt.legend()
plt.show()
```
